## Importing libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np  
import seaborn as sns
import warnings
from warnings import filterwarnings
filterwarnings('ignore')
# %matplotlib inline
```

```python
df=pd.read_csv('Datasets/kc_house_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
df['date']=pd.to_datetime(df['date'])
df['month']=df['date'].apply(lambda date:date.month)
df['year']=df['date'].apply(lambda date:date.year)
df.head()
```

```python
df.drop('date',axis=1,inplace=True) 
```

```python
df.head()
```

```python
df.drop('zipcode',axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler=MinMaxScaler()

X=df.drop('price',axis=1).values
y=df['price'].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

## Model Training

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')

model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score

predictions=model.predict(X_test)

mean_absolute_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.scatter(y_test, predictions)
```

## Retraning the model

```python
new_X=df.drop('price',axis=1).values
new_y=df['price'].values

new_X_train,new_X_test,new_y_train,new_y_test=train_test_split(new_X,new_y,test_size=0.2,random_state=101)

new_X_train=scaler.fit_transform(new_X_train)
new_X_test=scaler.transform(new_X_test)

model=Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')

model.fit(x=new_X_train,y=new_y_train,validation_data=(new_X_test,new_y_test),batch_size=128,epochs=400)
```

```python
new_loss=pd.DataFrame(model.history.history)

new_loss.plot()
```

## Prediction and Evaluation

```python
model.predict(new_X_test)

mean_absolute_error(new_y_test,model.predict(new_X_test))
```

```python
np.sqrt(mean_squared_error(new_y_test,model.predict(new_X_test)))
```

```python
explained_variance_score(new_y_test,model.predict(new_X_test))
```
